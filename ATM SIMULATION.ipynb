```python
correct_pin = "1234"
balance = 10000
attempts = 0

while attempts < 3:
    pin = input("Enter ATM PIN: ")

    if pin == correct_pin:
        print("\nAccess Granted! Welcome to your account.")
        print("Current Balance:", balance)

        # Withdrawal
        try:
            amount = int(input("Enter the amount to withdraw: "))
        except ValueError:
            print("Invalid amount! Please enter numbers only.")
            continue

        if amount > balance:
            print("Insufficient Balance! Transaction failed.")
        else:
            balance -= amount
            print("Transaction Successful!")
            print("Remaining Balance:", balance)

        break  # Exit after successful login

    else:
        attempts += 1
        print(f"Wrong PIN! Attempts left: {3 - attempts}")

# Block ATM after 3 failed attempts
if attempts == 3:
    print("\nYour ATM has been blocked due to multiple wrong attempts!")
```

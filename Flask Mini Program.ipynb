```python
from flask import Flask, request

app = Flask(_name_)

items = []

@app.route("/")
def home():
    return "Simple Flask API is running!"

@app.route("/add", methods=["POST"])
def add_item():
    item = request.json["item"]
    items.append(item)
    return f"Added: {item}"

@app.route("/items")
def get_items():
    return {"items": items}

if _name_ == "_main_":
    app.run(debug=True)
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((600, 600))
pygame.display.set_caption("X O Game")

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
size = 200

# Reset function
def reset_game():
    return [["", "", ""],
            ["", "", ""],
            ["", "", ""]], "X", False, ""

board, player, game_over, winner = reset_game()

font = pygame.font.Font(None, 100)
small_font = pygame.font.Font(None, 50)

run = True
while run:
    screen.fill(WHITE)

    # Draw grid lines
    pygame.draw.line(screen, BLACK, (200, 0), (200, 600), 5)
    pygame.draw.line(screen, BLACK, (400, 0), (400, 600), 5)
    pygame.draw.line(screen, BLACK, (0, 200), (600, 200), 5)
    pygame.draw.line(screen, BLACK, (0, 400), (600, 400), 5)

    # Draw X and O
    for row in range(3):
        for col in range(3):
            if board[row][col] != "":
                text = font.render(board[row][col], True, BLACK)
                screen.blit(text, (col * size + 70, row * size + 50))

    # Check events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        
        # Restart when ENTER is pressed
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RETURN:
                board, player, game_over, winner = reset_game()

        # Mouse click to place X or O
        if event.type == pygame.MOUSEBUTTONDOWN and not game_over:
            mouse_x, mouse_y = pygame.mouse.get_pos()
            row = mouse_y // 200
            col = mouse_x // 200

            if board[row][col] == "":
                board[row][col] = player

                # Check win
                # Rows
                for r in board:
                    if r == [player, player, player]:
                        winner = player
                        game_over = True

                # Columns
                for c in range(3):
                    if (board[0][c] == player and
                        board[1][c] == player and
                        board[2][c] == player):
                        winner = player
                        game_over = True

                # Diagonals
                if (board[0][0] == player and board[1][1] == player and board[2][2] == player):
                    winner = player
                    game_over = True

                if (board[0][2] == player and board[1][1] == player and board[2][0] == player):
                    winner = player
                    game_over = True

                # Check draw
                if not game_over:
                    full = True
                    for r in board:
                        if "" in r:
                            full = False
                    if full:
                        winner = "Draw"
                        game_over = True

                # Switch player
                if not game_over:
                    player = "O" if player == "X" else "X"

    # Display winner or draw
    if game_over:
        msg = "Draw!" if winner == "Draw" else f"{winner} Wins!"
        text = small_font.render(msg + "  Press ENTER to restart.", True, BLACK)
        screen.blit(text, (50, 560))

    pygame.display.update()

pygame.quit()
```

```python
name = input("Enter Your Name: ")
age = int(input("Enter Your Age: "))

if age < 0:
    print("Invalid Age")

elif age > 0 and age < 18:
    print("You Can't Book the ticket Due to being minor")

elif age >= 18 and age < 60:
    print("You can book your ticket")
    print(
        """
        Select the Slot:
        1. 26/11/2025 09:00 AM TO 01:00 PM
        2. 27/11/2025 09:00 AM TO 01:00 PM
        3. 28/11/2025 09:00 AM TO 01:00 PM
        """
    )
    slot = int(input("Enter your slot choice: "))

    if 1 <= slot <= 3:
        print("Ticket Booked Successfully!")
    else:
        print("Invalid Slot Selection!")

else:
    print("You can book Ticket with Guardian")
    choice1 = int(input("Wish to continue? Press 1 to continue or 2 to exit: "))

    if choice1 == 1:
        gname = input("Enter Your Guardian Name: ")
        print(
            """
            Select the Slot:
            1. 26/11/2025 09:00 AM TO 01:00 PM
            2. 27/11/2025 09:00 AM TO 01:00 PM
            3. 28/11/2025 09:00 AM TO 01:00 PM
            """
        )
        slot = int(input("Enter your slot choice: "))

        if 1 <= slot <= 3:
            print(f"Ticket Booked Successfully with guardian {gname}!")
        else:
            print("Invalid Slot Selection!")
    else:
        print("Thank you! Visit Again.")
```
